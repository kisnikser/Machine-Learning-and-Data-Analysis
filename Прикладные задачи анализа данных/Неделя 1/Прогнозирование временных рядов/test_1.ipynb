```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
milk.head()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(milk.milk)
```

```python
import statsmodels.api as sm
sm.tsa.stattools.adfuller(milk.milk)
```

```python
days = [milk.index[i].days_in_month for i in range(milk.shape[0])]
```

```python
days[1]
```

```python
milk_per_day = []

for i in range(milk.shape[0]):
    milk_per_day.append(milk.milk[i] / days[i])
```

```python
milk_per_day
```

```python
plt.figure(figsize = (10, 6))
plt.plot(milk_per_day)
```

```python
print(sum(milk_per_day))
```
